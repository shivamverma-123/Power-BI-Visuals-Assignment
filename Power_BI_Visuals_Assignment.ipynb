Question 3 : Create a Line Chart showing the Sales Trend over time (by Month and Year). Is there any
noticeable seasonal pattern in sales?

Ans:The line chart showing the sales trend over time by months and year are

Step 1: Insert Visual

- Go to Visualizations pane

- Select Line Chart

 Step 2: Add Fields

- Axis → Order Date

- Click dropdown → Select Date Hierarchy

- Keep Year and Month

- Values → Sales

- Aggregation: Sum

Step 3: Format (Optional but Recommended)

- Turn ON Data Labels

- X-axis → Show as Continuous

- Title: Sales Trend by Month & Year.
